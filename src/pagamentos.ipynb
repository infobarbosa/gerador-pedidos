```python
filenames = [
'pedidos-2024-01-04.txt',  'pedidos-2024-01-09.txt',  'pedidos-2024-01-14.txt',  'pedidos-2024-01-19.txt',  'pedidos-2024-01-24.txt',  'pedidos-2024-01-29.txt',
'pedidos-2024-01-05.txt',  'pedidos-2024-01-10.txt',  'pedidos-2024-01-15.txt',  'pedidos-2024-01-20.txt',  'pedidos-2024-01-25.txt',  'pedidos-2024-01-30.txt',
'pedidos-2024-01-01.txt',  'pedidos-2024-01-06.txt',  'pedidos-2024-01-11.txt',  'pedidos-2024-01-16.txt',  'pedidos-2024-01-21.txt',  'pedidos-2024-01-26.txt',  
'pedidos-2024-01-02.txt',  'pedidos-2024-01-07.txt',  'pedidos-2024-01-12.txt',  'pedidos-2024-01-17.txt',  'pedidos-2024-01-22.txt',  'pedidos-2024-01-27.txt',
'pedidos-2024-01-03.txt',  'pedidos-2024-01-08.txt',  'pedidos-2024-01-13.txt',  'pedidos-2024-01-18.txt',  'pedidos-2024-01-23.txt',  'pedidos-2024-01-28.txt',
'pedidos-2024-01-31.txt'
]

bigdoc = json.loads("[]")
for file in filenames:
    with open(f'../data/pedidos/{file}', 'r') as f:
        for linha in f:
            doc = json.loads(linha)
            print(doc["id_pedido"] + ", id_cliente " + str(doc["id_cliente"]))
            
            status_pedido = True
            indicador_fraude = False

            if doc["id_cliente"] % 13 == 0:
                status_pedido = False
                indicador_fraude = True
            elif doc["id_pedido"][-1] == "d" or doc["id_pedido"][-1] == "9":
                status_pedido = False
            new_doc = {
                "id_pedido": doc["id_pedido"],
                "status": status_pedido,
                "fraude": indicador_fraude
            }

            bigdoc.append(new_doc)
```
